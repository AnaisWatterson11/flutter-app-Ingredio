```python
import pandas as pd

# Chargement du fichier
file_path1 = "recettes_sans_doublons.csv"
df1 = pd.read_csv(file_path1)

# Vérifier les valeurs manquantes (NaN)
missing_values = df1.isnull().sum()
print("Nombre de valeurs manquantes par colonne :")
print(missing_values)

# Vérifier les lignes où certaines colonnes contiennent "Non spécifié"
colonnes_a_verifier = ["Coût total", "Coût par part", "Temps de préparation", "Temps de cuisson"]

# Filtrer les lignes où au moins une de ces colonnes contient "Non spécifié"
non_specifie = df1[colonnes_a_verifier].apply(lambda x: x == "Non spécifié").any(axis=1)

# Compter et afficher le nombre de lignes concernées
nombre_non_specifie = non_specifie.sum()
print(f"\nNombre de lignes contenant 'Non spécifié' dans les colonnes ciblées : {nombre_non_specifie}")

# (Optionnel) Voir les lignes concernées
df_non_specifie = df1[non_specifie]
display(df_non_specifie)
```

```python
print(f"Le fichier a {df1.shape[0]} lignes et {df1.shape[1]} colonnes.")
```

```python
df1.shape
```

```python
# Supprimer les lignes contenant 'Non spécifié'
df1_cleaned = df1[~non_specifie]

# Sauvegarder dans un nouveau fichier CSV
df1_cleaned.to_csv("recettes_finales.csv", index=False)

print("Le fichier nettoyé a été sauvegardé sous le nom 'recettes_finales.csv'.")
```

```python
df1_cleaned.shape
```

```python
df1_cleaned
```

```python
# Vérifier les doublons sur les colonnes "Nom" et "Catégorie"
doublons = df1_cleaned.duplicated(subset=["Nom", "Catégorie"])

# Nombre de doublons trouvés
nombre_doublons = doublons.sum()
print(f"Nombre de recettes dupliquées selon 'Nom' et 'Catégorie' : {nombre_doublons}")

# (Optionnel) Voir les lignes dupliquées
df_doublons = df1_cleaned[doublons]
display(df_doublons)
```
