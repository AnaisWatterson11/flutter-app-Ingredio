```python
import pandas as pd

# Charger le fichier CSV
file_path = "recettes_extractees.csv"
df = pd.read_csv(file_path)

# Afficher la dimension initiale du fichier
print(f"Dimensions initiales: {df.shape}")
```

```python

# Vérifier les valeurs manquantes par colonne
missing_values = df.isnull().sum()

# Afficher les colonnes avec des valeurs manquantes
print("\nValeurs manquantes par colonne :")
print(missing_values)
```

```python
# Supprimer les lignes avec des valeurs manquantes dans 'Ingrédients' ou 'Étapes'
df_cleaned = df.dropna(subset=["Ingrédients", "Étapes"])

# Afficher la dimension du fichier après nettoyage
print(f"\nDimensions après suppression des valeurs manquantes : {df_cleaned.shape}")
```

```python

# Vérifier les valeurs manquantes par colonne
missing_values2 = df_cleaned.isnull().sum()

# Afficher les colonnes avec des valeurs manquantes
print("\nValeurs manquantes par colonne :")
print(missing_values2)
```

```python
# Afficher les lignes avec des valeurs manquantes
missing_rows1 = df_cleaned[df_cleaned.isnull().any(axis=1)]
print("\nLignes avec des valeurs manquantes :")
print(missing_rows1)
```

```python
# Vérifier les recettes redondantes dans le même mode
duplicate_within_mode = df_cleaned[df_cleaned.duplicated(subset=["Nom", "Mode"], keep=False)]

# Afficher les recettes redondantes dans le même mode
print("\nRecettes redondantes dans le même mode :")
print(duplicate_within_mode)
```

```python
# Vérifier les recettes redondantes pour des modes différents
duplicate_across_modes = df_cleaned[df_cleaned.duplicated(subset=["Nom"], keep=False)]

# Afficher les recettes redondantes pour des modes différents
print("\nRecettes redondantes pour des modes différents :")
print(duplicate_across_modes)
```

```python

# Vérifier les valeurs uniques pour chaque colonne
for column in df_cleaned.columns:
    print(f"\nValeurs possibles pour la colonne '{column}':")
    print(df_cleaned[column].unique())

    # Vérifier s'il y a la valeur "Non spécifié" dans la colonne
    if "Non spécifié" in df_cleaned[column].values:
        print(f"\nIl y a des occurrences de 'Non spécifié' dans la colonne '{column}'")
    else:
        print(f"\nIl n'y a pas de 'Non spécifié' dans la colonne '{column}'")
```

```python
# Supprimer les lignes où la colonne 'Calories' contient la valeur 'Non spécifié'
df_cleaned2 = df_cleaned[df_cleaned['Calories'] != 'Non spécifié']

# Afficher la nouvelle dimension du DataFrame après suppression
print(f"Dimensions après suppression : {df_cleaned2.shape}")

# Sauvegarder le fichier nettoyé
df_cleaned2.to_csv("recettes_SM2.csv", index=False)
```

```python
import pandas as pd

# Charger le fichier recettes_SM2.csv
file_path_SM2 = "recettes_SM2.csv"
df_SM2 = pd.read_csv(file_path_SM2)

# Vérifier les recettes redondantes pour le même mode (même Nom, Mode et Catégorie)
duplicate_within_mode = df_SM2[df_SM2.duplicated(subset=["Nom", "Mode", "Catégorie"], keep=False)]
print("\nRecettes redondantes pour le même mode :")
print(duplicate_within_mode)
```

```python
# Supprimer les doublons et ne garder que la première occurrence
df_SM2_cleaned = df_SM2.drop_duplicates(subset=["Nom", "Mode", "Catégorie"], keep="first")

# Afficher la dimension du DataFrame après suppression des doublons
print(f"\nDimensions après suppression des doublons : {df_SM2_cleaned.shape}")
```

```python
# Sauvegarder dans un nouveau fichier CSV
df_SM2_cleaned.to_csv("recettes_SM2_sans_redondance.csv", index=False)

print("\nVérifications terminées.")
```
